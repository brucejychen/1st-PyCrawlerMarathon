### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import datetime
import re
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

pattern ='[0-9]{4}(?:/[0-9]{2}){2}'
today = datetime.date.today().strftime("%Y/%m/%d")

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all("h3"):
    if re.match(pattern, d.find(class_="date").text)[0] == today:
        print(d.find(class_="date").text,d.find_all('a')[-1].text )
    

    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.em.string,d.a.text)

```

### ③ 哪一個類別的文章最多

```python
article=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    article.append([d.find('em').text, d.find_all('a')[-1].text])
    
columns=['類別','標題']
import pandas as pd

df= pd.DataFrame(article, columns=columns)

df2= pd.DataFrame(df.groupby('類別').size(),columns=['數量'])

print(df2[df2['數量']==max(df2['數量'])])
```
